```python
uygulama=pd.DataFrame(
    {
        "Tarih":pd.date_range("20000101",periods=276,freq="m"),
        "Sıcaklık":pd.Series(np.random.randint(0,38, size=276)),
        "Yağış Oranı":pd.Series(np.random.random(276),dtype="float32"),
    }
)
uygulama.describe()
```

```python
x = ['Karlı','Yağmurlu','Bulutlu','Parcali Bulutlu','Günesli']
cut_bins = [0,3,9,17,24,38]
```

```python
uygulama['Hava_Durumu'] = pd.cut(uygulama['Sıcaklık'],bins = cut_bins, labels = x)
```
